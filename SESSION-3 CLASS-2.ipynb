# DATA MODEL = STAR SCHEMA

```python
step_1 = pd.merge(left=transaction_master, right=customer_master, how='left', on='CUST_ID')
step_1
```

```python
step_2 = pd.merge(left=step_1, right = product_master, how = 'left', on = 'PROD_ID')
step_3 = pd.merge(left=step_2, right=store_master, how ='left', on = 'STORE_ID')
data_model_output = pd.merge(left=step_3, right=staff_master, how = 'left', on = 'STAFF_ID')
```

```python
data_model_output
```

# Data Model = Snow Flake

```python
Snowflake_schema = pd.merge(left=data_model_output, right=supplier_master, how = 'left', on= 'SUPPLIER_ID')
Snowflake_schema
```
