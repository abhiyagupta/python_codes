# Ranking in Python

```python
med_summary['Rank'] = med_summary['SUBS'].rank()

med_summary
```

```python
med_summary['Rank2'] = med_summary['SUBS'].rank(ascending = False)

med_summary
```

# Cumulative Sum

```python
med_summary=med_2022.groupby(['STATE_CODE','Company'],as_index=False).agg(SUBS=('CUSTOMER_ID','count'))
med_summary
```

```python
med_summary['CUM_SUM_SUBS']=med_summary.groupby(['STATE_CODE'])['SUBS'].apply(lambda x: x.cumsum())

med_summary
```
