# MERGING TABLES

```python
# 2. HORIZONTAL JOIN (MERGING TABLES)

import pandas as pd
import numpy as np

stu_education=pd.read_excel(r'C:\Users\Debendra\BATCH 202209\SESSION-3\A2.RAW_DATA\A3.HORIZONTAL JOIN DATA-BASIC\A3.HORIZONTAL JOIN DATA.xlsx',sheet_name='STU_EDUCATION')
stu_experience=pd.read_excel(r'C:\Users\Debendra\BATCH 202209\SESSION-3\A2.RAW_DATA\A3.HORIZONTAL JOIN DATA-BASIC\A3.HORIZONTAL JOIN DATA.xlsx',sheet_name='STU_EXPERIENCE')

print(stu_education)
print(stu_experience)

```

```python
# INNER JOIN

stu_full_details=pd.merge(left=stu_education,right=stu_experience,how='inner',on='STU_ID')

stu_full_details
```

```python
# FULL JOIN

stu_full_details=pd.merge(left=stu_education,right=stu_experience,how='outer',on='STU_ID')

stu_full_details
```

```python
# LEFT JOIN

stu_full_details=pd.merge(left=stu_education,right=stu_experience,how='left',on='STU_ID')

stu_full_details
```

```python
# RIGHT JOIN

stu_full_details=pd.merge(left=stu_education,right=stu_experience,how='right',on='STU_ID')

stu_full_details
```

```python
# NULL JOIN

# LEFT NULL JOIN

stu_left_details=pd.merge(left=stu_education,right=stu_experience,how='left',on='STU_ID')
stu_left_null_details=stu_left_details[stu_left_details['COMPANY'].isnull()]
stu_left_null_details
```

```python
# NULL JOIN

# RIGHT NULL JOIN

stu_right_details=pd.merge(left=stu_education,right=stu_experience,how='right',on='STU_ID')
stu_right_null_details=stu_right_details[stu_right_details['STU_NAME'].isnull()]
stu_right_null_details
```

```python
# NULL JOIN

# UN_MATCHED JOIN

stu_unmatched_details=pd.concat([stu_left_null_details,stu_right_null_details],ignore_index=True)
stu_unmatched_details

```

```python
# OR

# NULL JOIN

# UN_MATCHED JOIN

stu_full_details=pd.merge(left=stu_education,right=stu_experience,how='outer',on='STU_ID')
stu_unmatched_details=stu_full_details[stu_full_details['STU_NAME'].isnull() | stu_full_details['COMPANY'].isnull()]
stu_unmatched_details
```

# JOINING DATASETS WHERE BOTH TABLES HAVING SAME FIELD NAME

```python
import pandas as pd

SALES_2019 = pd.DataFrame({'PRODUCT':['APPLE','DELL','HP','ACER','LENOVO'],
                          'YEAR':[2019,2019,2019,2019,2019],
                          'SALES':[130,230,340,890,900]})
SALES_2019

SALES_2020 = pd.DataFrame({'PRODUCT':['APPLE','DELL','HP','ACER','LENOVO'],
                          'YEAR':[2020,2020,2020,2020,2020],
                          'SALES':[789,120,310,780,990]})
SALES_2020

print(SALES_2019)
print(SALES_2020)
```

```python
sales_compare=pd.merge(left=SALES_2019,right=SALES_2020,how='inner',on='PRODUCT')
sales_compare
```

```python
sales_compare=pd.merge(left=SALES_2019,right=SALES_2020,how='inner',on='PRODUCT',suffixes=('_left','_right'))
sales_compare['delta']=(1-(sales_compare['SALES_left']/sales_compare['SALES_right'])).map(lambda n: '{:,.2%}'.format(n))
sales_compare
```

# SUB-QUERIES

```python
import pandas as pd
import numpy as np

file_path=r'C:\Users\Debendra\BATCH_202207_PYTHON\SESSION-3\A2.RAW_DATA\A5.HORIZONTAL JOIN DATA-ADVANCE\A4.HORIZONTAL JOIN DATA.xlsx'

prod_sales=pd.read_excel(file_path,sheet_name='SUB_QUERY_DATA1')

prod_sales_summary=prod_sales.groupby(['COUNTRY'],as_index=False).agg(total_sales=('UNITS_SOLD','sum'))

# print(prod_sales_summary)
# print(prod_sales)

prod_sales_market_share=pd.merge(left=prod_sales,right=prod_sales_summary,how='outer',on='COUNTRY')
prod_sales_market_share['market_share']=(prod_sales_market_share['UNITS_SOLD']/prod_sales_market_share['total_sales']).map(lambda n: '{:,.2%}'.format(n))
prod_sales_market_share
```

# JOINING AND RELATION IN PYTHON PROGRAMMING

```python
import pandas as pd
import numpy as np

file_path=r'C:\Users\Debendra\BATCH 202209\SESSION-3\A2.RAW_DATA\A6.RELATIONSHIP DATA\A5.RELATIONSHIP DATA.xlsx'

players_score=pd.read_excel(file_path,sheet_name='PLAYERS_SCORE')
players_details=pd.read_excel(file_path,sheet_name='PLAYERS_DETAILS')
match_score=pd.read_excel(file_path,sheet_name='MATCH_SCORE')
series_score=pd.read_excel(file_path,sheet_name='SERIES_SCORE')
players_score_cont=pd.read_excel(file_path,sheet_name='PLAYERS_SCORE_CONTRIBUTION')

print(players_score)
print(players_details)
print(match_score)
print(series_score)
print(players_score_cont)
```

# RELATIONSHIP

```python
# RELATIONSHIP ARE 4 TYPES
# 1. ONE TO ONE
# 2. ONE TO MANY
# 3. MANY TO ONE
# 4. MANY TO MANY
```

```python
# 1. ONE TO ONE
# print(players_score)
# print(players_details)

players_full_details=pd.merge(left=players_score,right=players_details,how='left',on='PLAYERS_ID')

players_full_details
```

```python
# 2. ONE TO MANY

# print(players_details)
# print(match_score)

players_match_details=pd.merge(left=players_details,right=match_score,how='left',on='PLAYERS_ID')
players_match_details
```

```python
# 3. MANY TO ONE
# print(match_score)
# print(players_details)

players_match_details=pd.merge(left=match_score,right=players_details,how='left',on='PLAYERS_ID')
players_match_details
```

```python
# 4. MANY TO MANY
# print(series_score)
# print(players_score_cont)

players_series_contribution=pd.merge(left=series_score,right=players_score_cont,how='outer',on='COUNTRY')
players_series_contribution['players_score']=players_series_contribution['TOTAL SCORE']*players_series_contribution['PROPORTION']
players_series_contribution
```

# DATA MODEL DESIGN USING PYTHON PROGRAMMING

```python
# import pandas as pd
import numpy as np

file_path=r'C:\Users\Debendra\BATCH 202209\SESSION-3\A2.RAW_DATA\A7.DATA MODEL IN PYTHON\A1.DATA MODEL BY EXAMPLE DATA.xlsx'

transaction_master=pd.read_excel(file_path,sheet_name='TRANSACTION_TABLE')
customer_master=pd.read_excel(file_path,sheet_name='CUSTOMER_TABLE')
product_master=pd.read_excel(file_path,sheet_name='PRODUCT_TABLE')
store_master=pd.read_excel(file_path,sheet_name='STORE_TABLE')
staff_master=pd.read_excel(file_path,sheet_name='STAFF_MASTER')
supplier_master=pd.read_excel(file_path,sheet_name='SUPPLIER_MASTER')


print(transaction_master)
print(customer_master)
print(product_master)
print(store_master)
print(staff_master)
print(supplier_master)
```

# DATA MODEL- STAR SCHEMA

```python
# print(transaction_master)
# print(customer_master)
# print(product_master)
# print(store_master)
# print(staff_master)

step_1=pd.merge(left=transaction_master,right=customer_master,how='left',on='CUST_ID')
step_2=pd.merge(left=step_1,right=product_master,how='left',on='PROD_ID')
step_3=pd.merge(left=step_2,right=store_master,how='left',on='STORE_ID')
data_model_output=pd.merge(left=step_3,right=staff_master,how='left',on='STAFF_ID')
data_model_output['SALES']=data_model_output['QTY']*data_model_output['PRICE']
data_model_output

data_model_output.to_excel(r'C:\Users\Debendra\BATCH 202209\SESSION-3\A4.OUTPUT\data_model_output.xlsx',index=None)

print(data_model_output)
```
