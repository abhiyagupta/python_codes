```python
prod_sales = pd.read_excel(r'C:\Users\Debendra das\Downloads\drive-download-20230506T083649Z-001\SESSION-2\A2.RAW_DATA\A3.NESTED CASE WHEN DATA.xlsx')
prod_sales
```

```python
def calculate_discount_price(row):
    product = row['PRODUCT']
    city = row['CITY']
    price = row['PRICE']
    
    if product == 'APPLE' :
        if city == "BANGALORE" :
            return 0.20*price
        elif city == 'CHENNAI' :
            return 0.30*price
        else:
            return 0
    elif product == 'DELL':
        if city == 'MUMBAI':
            return 0.20*price
        elif city == 'DELHI':
            return 0.30*price
        else :
            return 0
    else :
        return 0

prod_sales['Discount'] = prod_sales.apply(calculate_discount_price, axis = 1)

prod_sales
```
